## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import numpy as np

def tirada_de_dados(n_simulación = 1000):
    conteo = 0
    
    # Cada iteración del bucle for es una prueba.
    for i in range(n_simulación):
        
        # Tirada de cada Dado
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        
        # Suma los valores para obtener la puntuación
        puntuación = dado1 + dado2
        
        # Decide si debemos agregarlo al conteo
        if (puntuación % 2 == 0) or (puntuación > 7):
            conteo += 1

    # Calculate probability
    return conteo / n_simulación
print(f"La probabilidad de obtener un número par o mayor que 7 es: {np.round(tirada_de_dados() * 100, 2)}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# 

def sacar_bolas(n_simulación = 1000):
    conteo_1 = 0
    conteo_2 = 0

    for i in range(n_simulación):
        colores = []

        # Take 5 balls from the box
        for i in range(5):
            colores.append(ball_box[np.random.randint(0, 59)])

        # Convert list to Numpy array for better filtering
        colores = np.array(colores)
        
        Bolas_Blancas = sum(colores == "White")
        Bolas_Rojas = sum(colores == "Red")
        Bolas_Verdes = sum(colores == "Green")

        # Decide if we should add it to the count:
        if (Bolas_Blancas == 3) and (Bolas_Rojas == 2):
            conteo_1 += 1
        
        if (Bolas_Blancas == 5) or (Bolas_Rojas == 5) or (Bolas_Verdes == 5):
            conteo_2 += 1
    
    return conteo_1 / n_simulación, conteo_2 / n_simulación

probabilidad = sacar_bolas(1000)

print(f"La probabilidad de que salgan 3 blancas y 2 rojas es: {np.round(probabilidad[0] * 100, 2)}%")
print(f"La probabilidad de que todas sean del mismo color es: {np.round(probabilidad[1] * 100, 2)}%")
```
